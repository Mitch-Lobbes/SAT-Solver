```python
import pandas
```

```python
DPLL = "results/Algorithm.DPLL.csv"
JEROSLOW = "results/Algorithm.JEROSLOW_WANG.csv"
DYNAMIC = "results/Algorithm.DYNAMIC_VARIABLE_ORDERING.csv"
```

```python
dpll_df = pandas.read_csv(DPLL)
jeroslow_df = pandas.read_csv(JEROSLOW)
dynamic_df = pandas.read_csv(DYNAMIC)
```

```python
import matplotlib.pyplot as plt
```

```python
dpll_means = dpll_df.groupby(['amount_vars']).mean()
dynamic_means = dynamic_df.groupby(['amount_vars']).mean()
jeroslow_means = jeroslow_df.groupby(['amount_vars']).mean()
```

```python
fig = plt.figure()
plt.plot(dynamic_means["solving_time"])
plt.plot(dpll_means["solving_time"])
plt.plot(jeroslow_means["solving_time"])

plt.show()
```

```python
ax = dpll_means["solving_time"].plot(label='DPLL') 
dynamic_means["solving_time"].plot(ax=ax, label='Dynamic variable ordering')
jeroslow_means["solving_time"].plot(ax=ax, label='Jeroslow-Wang')
plt.legend()
plt.savefig('solving_time.svg')
```

```python
ax = dpll_means["avg_backtracks"].plot(label='DPLL') 
dynamic_means["avg_backtracks"].plot(ax=ax, label='Dynamic variable ordering')
jeroslow_means["avg_backtracks"].plot(ax=ax, label='Jeroslow-Wang')
plt.legend()
plt.savefig('backtracks.svg')
```

```python
ax = dpll_means["avg_splits"].plot(label='DPLL') 
dynamic_means["avg_splits"].plot(ax=ax, label='Dynamic variable ordering')
jeroslow_means["avg_splits"].plot(ax=ax, label='Jeroslow-Wang')
plt.legend()
plt.savefig('splits.svg')
```
